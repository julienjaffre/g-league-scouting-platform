```python
import pandas as pd
import os
from IPython.display import display, HTML

# ðŸ“ Chemin vers les fichiers bronze
bronze_dir = "../data/bronze"

# ðŸ” Charger tous les fichiers CSV contenant "_2021_2024" ou "_2022_2024"
bronze_files = {
    os.path.splitext(f)[0]: os.path.join(bronze_dir, f)
    for f in os.listdir(bronze_dir)
    if f.endswith(".csv") and ("_2021_2024" in f or "_2022_2024" in f or "_2023_2024" in f)
}

# ðŸ“Š Chargement
dfs = {name: pd.read_csv(path) for name, path in bronze_files.items()}

# ðŸ“‹ Affichage des 10 premiÃ¨res lignes de chaque table + saisons si disponibles
for name, df in dfs.items():
    print(f"\nðŸ”· Table `{name}` â€” {df.shape[0]} lignes, {df.shape[1]} colonnes")
    display(HTML(df.head(10).to_html(index=False)))

    # ðŸ”Ž VÃ©rification des saisons si la colonne "season" existe
    if "season" in df.columns:
        saisons = df["season"].dropna().astype(str).unique()
        print(f"ðŸ“† Saisons prÃ©sentes dans `{name}` : {sorted(saisons)}")
```

```python
import os

def afficher_taille_fichiers(dossier):
    print(f"\nðŸ“¦ Tailles des fichiers CSV dans : {dossier}\n")
    for fichier in os.listdir(dossier):
        if fichier.endswith(".csv"):
            chemin = os.path.join(dossier, fichier)
            taille_bytes = os.path.getsize(chemin)
            taille_mo = taille_bytes / (1024 * 1024)
            print(f"ðŸ“ {fichier:<40} â€” {taille_mo:.2f} Mo")

# ðŸ” Appel
afficher_taille_fichiers("../data/bronze")
```
